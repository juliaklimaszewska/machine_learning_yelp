```python
import pandas as pd

# reading in dataset
yelp_business = pd.read_csv("yelp_business.csv")

# printing first five rows
print(yelp_business.head(n = 5))
```

```python
# get rows that contain the categorie "Bars" and "Nightlife"
yelp = yelp_business['categories'].str.contains('Bars', 'Nightlife')
yelp_df = yelp_business[yelp]
yelp_df.shape
```

```python
# removing the 'Sushi Bars' that don't contain 'Bars' as well
yelp_keep = yelp_df['categories'].str.contains('Tapas Bars' and 'Nightlife')
yelp_df = yelp_df[yelp_keep]
yelp_df.shape
```
